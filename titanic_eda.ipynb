```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Dataset shape:", df.shape)
df.head()
```

```python
print("Missing values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Sex', hue='Survived')
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(['Not Survived', 'Survived'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Pclass', hue='Survived')
plt.title('Survival by Ticket Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.legend(['Not Survived', 'Survived'])
plt.show()
```

```python
plt.figure(figsize=(8,5))
df['Age'].hist(bins=30, color='steelblue', edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
print("Survival rate by gender (%):")
print(round(df.groupby('Sex')['Survived'].mean() * 100, 1))

print("\nSurvival rate by class (%):")
print(round(df.groupby('Pclass')['Survived'].mean() * 100, 1))

print("\nOverall survival rate (%):")
print(round(df['Survived'].mean() * 100, 1))
```
